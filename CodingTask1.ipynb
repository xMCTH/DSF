```python
import pandas as pd
from google.colab import drive
import re
import os
import numpy as np

drive.mount('/content/drive')
path = '/content/drive/MyDrive/Colab Notebooks/CCD-Data.txt'

def remove_header(file_path):
    with open(path, 'r') as input_file:
      contents = re.split('>.*\n', input_file.read())     #The regular expression '>.*\n' matches any string that starts with a greater-than sign (>) followed by zero or more characters (.*) and ending with a newline character (\n) (CHATGPT)

    with open(path, 'w') as output_file:
      output_file.write(contents[-1])

remove_header(path)  


```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CCD-Data.txt', sep='\t')
df.head()
```

```python
def merge_columns(dataframe):
    renamed_columns = dataframe.rename(columns=lambda column_name: column_name.split('.')[0])
    grouped_columns = renamed_columns.iloc[:, 2:].groupby(level=0, axis=1).mean()
    return grouped_columns

```

```python
merching = merge_columns(df)
merching.head()
```

```python
def merge_columns_foreachnm(dataframe,nm):
    renamed_columns = dataframe.rename(columns=lambda column_name: column_name.split('.')[0])
    grouped_columns = renamed_columns.iloc[:, 2:].groupby(level=0, axis=1).mean()
    grouped_columns.columns = [nm] + grouped_columns.columns.tolist()
    
    return grouped_columns
```
